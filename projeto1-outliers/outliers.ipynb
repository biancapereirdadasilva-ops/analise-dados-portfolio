# Detecção de Outliers com Python

Este notebook demonstra como identificar valores discrepantes (outliers) em um conjunto de dados.

## 1. Criando uma lista de idades com outlier

```python
import numpy as np
import matplotlib.pyplot as plt

# Lista com outlier (400 é um valor discrepante)
lista_idades = [26, 30, 32, 22, 26, 35, 400, 20, 43, 31, 23]
print("Lista de idades:", lista_idades)
```

## 2. Calculando a média com e sem outlier

```python
# Média com outlier
media_com_outlier = np.mean(lista_idades)
print(f"Média com outlier: {media_com_outlier:.2f}")

# Removendo o outlier (400)
lista_sem_outlier = [26, 30, 32, 22, 26, 35, 20, 43, 31, 23]
media_sem_outlier = np.mean(lista_sem_outlier)
print(f"Média sem outlier: {media_sem_outlier:.2f}")
```

## 3. Calculando o desvio padrão

```python
desvio_com_outlier = np.std(lista_idades)
print(f"Desvio padrão com outlier: {desvio_com_outlier:.2f}")

desvio_sem_outlier = np.std(lista_sem_outlier)
print(f"Desvio padrão sem outlier: {desvio_sem_outlier:.2f}")
```

## 4. Boxplot para visualizar o outlier

```python
plt.figure(figsize=(8, 4))
plt.boxplot(lista_idades)
plt.title('Boxplot - Idades com Outlier')
plt.ylabel('Idade')
plt.grid(True, alpha=0.3)
plt.show()
```

## 5. Método IQR para identificar limites

```python
# Calculando quartis
Q1 = np.percentile(lista_sem_outlier, 25)
Q3 = np.percentile(lista_sem_outlier, 75)
IQR = Q3 - Q1

limite_superior = Q3 + (1.5 * IQR)
limite_inferior = Q1 - (1.5 * IQR)

print(f"Q1 (primeiro quartil): {Q1}")
print(f"Q3 (terceiro quartil): {Q3}")
print(f"IQR (intervalo interquartil): {IQR}")
print(f"Limite superior: {limite_superior}")
print(f"Limite inferior: {limite_inferior}")
```

## 6. Conclusão

- Outliers distorcem a média e o desvio padrão
- O boxplot ajuda a identificá-los visualmente
- O método IQR fornece limites objetivos
- Para dados com outliers, a mediana é mais confiável que a média
