```python
# Feature Engineering com Python

import pandas as pd

# 1. Dados de exemplo
dados = pd.DataFrame({
    'NOME': ['Ana', 'Bruno', 'Carla', 'Diego', 'Elena', 'Felipe', 'Gabriela'],
    'IDADE': [25, 42, 31, 19, 55, 28, 45],
    'GESTOR?': [0, 1, 0, 0, 1, 0, 1],
    'NIVEL': ['Junior', 'Senior', 'Pleno', 'Junior', 'Senior', 'Pleno', 'Senior']
})

print("📊 Dados originais:")
print(dados)

# ============================================
# TÉCNICA 1: Variável Condicional
# ============================================
print("\n" + "="*50)
print("TÉCNICA 1: Criando variável condicional")
print("="*50)

def preencher_nivel(gestor, nivel):
    if gestor == 1:
        return "Pessoa Gestora"
    return nivel

dados['NOVO_NIVEL'] = dados.apply(
    lambda x: preencher_nivel(x['GESTOR?'], x['NIVEL']),
    axis=1
)

print("\n✅ Com NOVO_NIVEL:")
print(dados[['NOME', 'GESTOR?', 'NIVEL', 'NOVO_NIVEL']])

# ============================================
# TÉCNICA 2: Variáveis Dummy (One-Hot Encoding)
# ============================================
print("\n" + "="*50)
print("TÉCNICA 2: Variáveis Dummy")
print("="*50)

dummies = pd.get_dummies(dados, columns=['NIVEL'])
print("\n✅ Com variáveis dummy:")
print(dummies)

# ============================================
# TÉCNICA 3: Discretização
# ============================================
print("\n" + "="*50)
print("TÉCNICA 3: Discretização (Idade → Geração)")
print("="*50)

def determinar_geracao(idade):
    if 39 < idade <= 58:
        return "Geração X"
    elif 29 < idade <= 39:
        return "Millennial"
    elif 13 < idade <= 29:
        return "Geração Z"
    else:
        return "Outra"

dados['GERACAO'] = dados['IDADE'].apply(determinar_geracao)

print("\n✅ Com coluna GERACAO:")
print(dados[['NOME', 'IDADE', 'GERACAO']])

# ============================================
# TÉCNICA 4: Criando faixas com pd.cut()
# ============================================
print("\n" + "="*50)
print("TÉCNICA 4: Criando faixas com pd.cut()")
print("="*50)

dados['FAIXA_ETARIA'] = pd.cut(dados['IDADE'],
                               bins=[0, 25, 35, 45, 100],
                               labels=['Até 25', '26-35', '36-45', '45+'])

print("\n✅ Com FAIXA_ETARIA:")
print(dados[['NOME', 'IDADE', 'FAIXA_ETARIA']])

# ============================================
# RESUMO FINAL
# ============================================
print("\n" + "="*50)
print("📊 RESUMO FINAL")
print("="*50)

print("\n📌 Técnicas aplicadas:")
print("1. Variável condicional (NOVO_NIVEL)")
print("2. Variáveis Dummy (One-Hot Encoding)")
print("3. Discretização (GERACAO)")
print("4. Criação de faixas (FAIXA_ETARIA)")

print(f"\n📊 Total de colunas antes: 4")
print(f"📊 Total de colunas depois: {len(dados.columns)}")

print("\n✅ Feature Engineering concluído!")
```
