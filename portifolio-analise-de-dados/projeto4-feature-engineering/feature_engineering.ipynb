# Feature Engineering com Python

Este notebook demonstra técnicas de criação de novas variáveis (features) a partir de dados existentes.

## 1. Importando bibliotecas

```python
import pandas as pd
import numpy as np
```

## 2. Criando dados de exemplo

```python
dados = pd.DataFrame({
    'NOME': ['Ana', 'Bruno', 'Carla', 'Diego', 'Elena'],
    'IDADE': [25, 42, 31, 19, 55],
    'GESTOR?': [0, 1, 0, 0, 1],
    'NIVEL': ['Junior', 'Senior', 'Pleno', 'Junior', 'Senior']
})

print("Dados originais:")
print(dados)
```

## 3. Técnica 1: Criando variável condicional

```python
def preencher_nivel(gestor, nivel):
    if gestor == 1:
        return "Pessoa Gestora"
    return nivel

dados['NOVO_NIVEL'] = dados.apply(
    lambda x: preencher_nivel(x['GESTOR?'], x['NIVEL']), axis=1
)

print("Com nova coluna 'NOVO_NIVEL':")
print(dados[['NOME', 'GESTOR?', 'NIVEL', 'NOVO_NIVEL']])
```

## 4. Técnica 2: Variáveis Dummy (One-Hot Encoding)

Transforma categorias em colunas binárias.

```python
dummies = pd.get_dummies(dados, columns=['NIVEL'])
print("Dados com variáveis dummy:")
print(dummies)
```

## 5. Técnica 3: Discretização (idade → geração)

```python
def determinar_geracao(idade):
    if 39 < idade <= 58:
        return "Geração X"
    elif 29 < idade <= 39:
        return "Millennial"
    elif 13 < idade <= 29:
        return "Geração Z"
    else:
        return "Outra"

dados['GERACAO'] = dados['IDADE'].apply(determinar_geracao)

print("Com coluna de geração:")
print(dados[['NOME', 'IDADE', 'GERACAO']])
```

## 6. Técnica 4: Criando faixas etárias

```python
dados['FAIXA_ETARIA'] = pd.cut(dados['IDADE'], 
                               bins=[0, 25, 35, 45, 100],
                               labels=['Até 25', '26-35', '36-45', '45+'])

print("Com faixa etária:")
print(dados[['NOME', 'IDADE', 'FAIXA_ETARIA']])
```

## 7. Comparação antes e depois

```python
print("=== ANTES ===")
print("Colunas:", list(dados.columns))
print(f"Total: {len(dados.columns)} colunas")

print("\n=== DEPOIS ===")
print("Colunas originais + novas:", list(dados.columns))
print(f"Total: {len(dados.columns)} colunas")
```

## 8. Conclusão

**Feature Engineering** é essencial porque:
- ✅ Transforma dados brutos em informações úteis
- ✅ Prepara dados para modelos de machine learning
- ✅ Cria variáveis que facilitam análises
- ✅ Ajuda a extrair padrões escondidos

**Técnicas aprendidas:**
1. Variáveis condicionais com `.apply()`
2. Variáveis dummy com `pd.get_dummies()`
3. Discretização com funções personalizadas
4. Criação de faixas com `pd.cut()`
