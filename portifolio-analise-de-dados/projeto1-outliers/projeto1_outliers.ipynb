```python
# Detecção de Outliers com Python

import numpy as np
import matplotlib.pyplot as plt

# 1. Lista com outlier
lista_idades = [26, 30, 32, 22, 26, 35, 400, 20, 43, 31, 23]
print("Lista de idades:", lista_idades)

# 2. Média com e sem outlier
media_com_outlier = np.mean(lista_idades)
print(f"Média com outlier: {media_com_outlier:.2f}")

lista_sem_outlier = [26, 30, 32, 22, 26, 35, 20, 43, 31, 23]
media_sem_outlier = np.mean(lista_sem_outlier)
print(f"Média sem outlier: {media_sem_outlier:.2f}")

# 3. Boxplot
plt.figure(figsize=(8, 4))
plt.boxplot(lista_idades)
plt.title('Boxplot - Idades com Outlier')
plt.ylabel('Idade')
plt.grid(True, alpha=0.3)
plt.show()

# 4. Método IQR
Q1 = np.percentile(lista_sem_outlier, 25)
Q3 = np.percentile(lista_sem_outlier, 75)
IQR = Q3 - Q1
limite_superior = Q3 + (1.5 * IQR)
limite_inferior = Q1 - (1.5 * IQR)

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Limite superior: {limite_superior}")
print(f"Limite inferior: {limite_inferior}")
```
