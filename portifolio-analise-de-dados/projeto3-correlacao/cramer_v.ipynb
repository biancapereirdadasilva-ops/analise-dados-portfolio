# Correlação de Variáveis Categóricas com Cramér's V

Este notebook demonstra como calcular a correlação entre variáveis categóricas usando o coeficiente de Cramér's V.

## 1. Importando bibliotecas

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
```

## 2. Criando dados de exemplo

```python
dados = pd.DataFrame({
    'COR_RACA': ['Branca', 'Parda', 'Branca', 'Preta', 'Parda', 'Branca', 'Amarela', 'Preta', 'Parda', 'Branca', 'Indígena', 'Parda'],
    'NIVEL_ENSINO': ['Superior', 'Médio', 'Pós', 'Superior', 'Médio', 'Superior', 'Médio', 'Pós', 'Superior', 'Médio', 'Superior', 'Pós']
})

print("Dados de exemplo:")
print(dados)
print(f"\nTotal de registros: {len(dados)}")
```

## 3. Tabela de Contingência (Crosstab)

Mostra a frequência com que os valores se cruzam.

```python
tabela_cruzada = pd.crosstab(dados['COR_RACA'], dados['NIVEL_ENSINO'])
print("Tabela de Contingência:")
print(tabela_cruzada)
```

## 4. Função Cramér's V

```python
def cramer_v(coluna1, coluna2):
    """
    Calcula o coeficiente de Cramér's V para medir associação entre variáveis categóricas.
    
    Parâmetros:
    coluna1, coluna2: pandas Series
    
    Retorna:
    float: coeficiente de Cramér's V (0 = sem associação, 1 = associação perfeita)
    """
    # Tabela de contingência
    tabela = pd.crosstab(coluna1, coluna2)
    
    # Teste qui-quadrado
    chi2 = chi2_contingency(tabela)[0]
    
    # Total de observações
    n = tabela.sum().sum()
    
    # Dimensão mínima da tabela - 1 (graus de liberdade)
    min_dim = min(tabela.shape) - 1
    
    # Cálculo do Cramér's V
    cramer = np.sqrt(chi2 / (n * min_dim))
    
    return cramer
```

## 5. Aplicando a função

```python
resultado = cramer_v(dados['COR_RACA'], dados['NIVEL_ENSINO'])
print(f"Cramér's V: {resultado:.4f}")
```

## 6. Interpretação do Resultado

```python
def interpretar_cramer(valor):
    if valor < 0.10:
        return "Associação muito fraca"
    elif valor < 0.20:
        return "Associação fraca"
    elif valor < 0.30:
        return "Associação moderada"
    elif valor < 0.50:
        return "Associação forte"
    else:
        return "Associação muito forte"

print(f"Resultado: {resultado:.4f}")
print(f"Interpretação: {interpretar_cramer(resultado)}")
```

## 7. Conclusão

- Cramér's V é ideal para **variáveis categóricas**
- Quanto mais próximo de 1, mais forte a associação
- Usa qui-quadrado como base para o cálculo
- Muito útil em análises exploratórias de dados
