```python
# Correlação de Variáveis Categóricas com Cramér's V

import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# 1. Função Cramér's V
def cramer_v(coluna1, coluna2):
    """
    Calcula o coeficiente de Cramér's V para medir associação entre variáveis categóricas.
    Retorna um valor entre 0 (sem associação) e 1 (associação perfeita).
    """
    # Tabela de contingência
    tabela = pd.crosstab(coluna1, coluna2)

    # Teste qui-quadrado
    chi2 = chi2_contingency(tabela)[0]

    # Total de observações
    n = tabela.sum().sum()

    # Graus de liberdade (menor dimensão - 1)
    min_dim = min(tabela.shape) - 1

    # Cálculo do Cramér's V
    cramer = np.sqrt(chi2 / (n * min_dim))

    return cramer

# 2. Dados de exemplo
dados = pd.DataFrame({
    'COR_RACA': ['Branca', 'Parda', 'Branca', 'Preta', 'Parda', 'Branca', 'Amarela', 'Preta', 'Parda', 'Branca', 'Indígena', 'Parda'],
    'NIVEL_ENSINO': ['Superior', 'Médio', 'Pós', 'Superior', 'Médio', 'Superior', 'Médio', 'Pós', 'Superior', 'Médio', 'Superior', 'Pós']
})

print("📊 Dados de exemplo:")
print(dados)
print(f"\nTotal de registros: {len(dados)}")

# 3. Tabela de contingência
tabela_cruzada = pd.crosstab(dados['COR_RACA'], dados['NIVEL_ENSINO'])
print("\n📈 Tabela de Contingência:")
print(tabela_cruzada)

# 4. Calcular correlação
resultado = cramer_v(dados['COR_RACA'], dados['NIVEL_ENSINO'])
print(f"\n🔗 Coeficiente de Cramér's V: {resultado:.4f}")

# 5. Interpretação
if resultado < 0.10:
    interpretacao = "Associação muito fraca"
elif resultado < 0.20:
    interpretacao = "Associação fraca"
elif resultado < 0.30:
    interpretacao = "Associação moderada"
elif resultado < 0.50:
    interpretacao = "Associação forte"
else:
    interpretacao = "Associação muito forte"

print(f"📝 Interpretação: {interpretacao}")

print("\n✅ Análise concluída!")
```
