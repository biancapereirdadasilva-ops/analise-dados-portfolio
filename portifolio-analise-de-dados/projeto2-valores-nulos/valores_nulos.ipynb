```python
# Tratamento de Valores Nulos com Pandas

import pandas as pd
import numpy as np

# 1. Criar dados de exemplo
dados = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5, 6, 7, 8],
    'SALARIO': [3000, 3500, None, 4000, None, 3200, 100000, 3800],
    'CARGO': ['Junior', 'Pleno', 'Junior', 'Senior', 'Pleno', 'Junior', 'CEO', 'Senior']
})

print("Dados originais:")
print(dados)

# 2. Identificar nulos
print(f"\nValores nulos por coluna:")
print(dados.isnull().sum())

# 3. Comparar média e mediana
media_salario = dados['SALARIO'].mean()
mediana_salario = dados['SALARIO'].median()

print(f"\nMédia: R$ {media_salario:.2f}")
print(f"Mediana: R$ {mediana_salario:.2f}")
print("✅ Usando mediana porque tem outlier (CEO)")

# 4. Preencher nulos com mediana
dados.loc[dados['SALARIO'].isnull(), 'SALARIO'] = mediana_salario

print("\nDados após preenchimento:")
print(dados)
```
